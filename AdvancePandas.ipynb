```python
from pandas_datareader import data, wb
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sales = pd.read_csv("salesUS.csv",index_col=1, parse_dates=True)
```

```python
sales.head()
```

```python
sales['TotSale'] = sales['Sales']*sales['Quantity']
```

```python
sales[['State', 'Profit', 'TotSale']]
```

```python
TX =sales[sales['State']== "Texas"]
TX
```

```python
TX_F = sales[(sales['State'] == "Texas") & (sales['Category'] == "Furniture")]
TX_F
```

```python
fig, ax = plt.subplots()
sales['Profit'].plot(ax=ax)
```

```python
fig, ax = plt.subplots()
TX_F['Profit'].plot(ax=ax, color='r', figsize = (14,6))
ax.grid()
ax.set_ylabel('Profit $')
ax.set_xlabel('Year / Month')
```

```python
fig, ax = plt.subplots(2)

sales['Profit'].loc[:'2011-1-31'].plot(ax=ax[0],title='Profit', color='b', figsize=(14,6))
sales['Sales'].loc[:'2011-1-31'].plot(ax=ax[1], title = 'Sales', color='r')
```

```python
fig, ax = plt.subplots()
sales['Profit'].loc['2011-12-01':'2011-12-05'].plot.bar(ax=ax, figsize = (14,6))
```

```python
s1 = sales.groupby(sales.index.year).sum()
s1
```

```python
fig, ax = plt.subplots()
s1['Sales'].plot.bar(ax=ax, figsize = (14,6))
```

```python
s2 = sales.groupby([sales.index.year, sales.index.month]).sum()
s2
```

```python
fig, ax = plt.subplots()
s2['Profit'].plot.bar(ax=ax, figsize = (14,6))
```

```python
fig, ax = plt.subplots()
s2[['Profit', 'Sales']].plot(ax=ax, figsize = (14,6))
ax.set_xlabel("Year/Month")
ax.set_ylabel("$")
ax.set_title("Sales vs Profit")
```

```python
central = sales[sales['Region']=="Central"]
south = sales[sales['Region']=="South"]
west = sales[sales['Region']=="West"]
east = sales[sales['Region']=="East"]
```

```python
sp_south = south.groupby([south.index.year, south.index.month]).sum()
sp_central = central.groupby([central.index.year,central.index.month]).sum()
sp_west = west.groupby([west.index.year, west.index.month]).sum()
sp_east = east.groupby([east.index.year, east.index.month]).sum()
```

```python
fig, ax = plt.subplots(2, 2)
sp_south[['Sales', 'Profit']].plot(ax=ax[0, 0], title="South Region",figsize = (14,6))
sp_central[['Sales', 'Profit']].plot(ax=ax[0, 1], title="Central Region",figsize = (14,6))
sp_east[['Sales', 'Profit']].plot(ax=ax[1, 0], title="East Region",figsize = (14,6))
sp_west[['Sales', 'Profit']].plot(ax=ax[1, 1], title="West Region",figsize = (14,6))

plt.tight_layout()
```
